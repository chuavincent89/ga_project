```python
import pandas as pd
import numpy as np
from datetime import datetime

import matplotlib.pyplot as plt
import seaborn as sns

from geopy import distance
```

```python
# load datasets
train = pd.read_csv("fraudTrain.csv")
test = pd.read_csv("fraudtest.csv")
```

```python
# check test datasets shape
print(test.shape)
test.head(3)
```

```python
# remove unuse first columns
train.drop(columns = train.columns[0], inplace=True)
test.drop(columns = test.columns[0], inplace=True)
```

```python
# change type to datetime for 'trans_date_trans_time' and 'dob'
train['trans_date_trans_time'] = pd.to_datetime(train['trans_date_trans_time'])
train['dob'] = pd.to_datetime(train['dob'])

test['trans_date_trans_time'] = pd.to_datetime(test['trans_date_trans_time'])
test['dob'] = pd.to_datetime(test['dob'])
```

```python
# change type of credit card number into object
train['cc_num'] = train['cc_num'].astype('object')
```

```python
# rename columns
train.rename(columns={'trans_date_trans_time':'trans_datetime'}, inplace=True)
test.rename(columns={'trans_datest_trans_time':'trans_datesttime'}, inplace=True)
```

```python
# combine first and last name
train['name'] = train['first'] + ' ' + train['last']

train.drop(columns=['first','last'], inplace=True)
```

```python
def conv_unix_time(unix):
    return datetime.datetime.fromtimestamp(unix).strftime('%Y-%m-%d %H:%M:%S')
```

```python
train['unix_time_conv'] = train['unix_time'].apply(conv_unix_time)
test['unix_time_conv'] = test['unix_time'].apply(conv_unix_time)
```

```python
train.columns
```

```python
# combine lat long
train['coords_ori'] = list(zip(train['lat'], train['long']))
train['coords_merch'] = list(zip(train['merch_lat'], train['merch_long']))
```

```python
train.head(2)
```

```python
def calculate_distance(ori, merch):
    return distance.distance(ori, merch).km
```

```python
train['distance'] = train.apply(lambda row: calculate_distance(row['coords_ori'], row['coords_merch']), axis=1)
```

```python
train.columns
```

```python
train['trans_datetime_year'] = train['trans_datetime'].dt.year
train['trans_datetime_month'] = train['trans_datetime'].dt.month
train['trans_datetime_week'] = train['trans_datetime'].dt.isocalendar().week 
train['trans_datetime_day'] = train['trans_datetime'].dt.day
train['trans_datetime_hour'] = train['trans_datetime'].dt.hour
train['trans_datetime_minute'] = train['trans_datetime'].dt.minute
train['trans_datetime_dayofweek'] = train['trans_datetime'].dt.dayofweek
```

```python
train.tail()
```

```python
train.info()
```

```python
train['dob_year'] = [year.year for year in train['dob']]
```

```python
train['age'] = train['trans_datetime_year'] - train['dob_year']
```

```python
cc_6 = train[train['cc_num'] == 60416207185].reset_index(drop=True)
```

```python
cc_6 = cc_6.drop(columns=['cc_num','street','city','state','zip','gender','lat','long','city_pop','job','dob','trans_num',
                   'unix_time','name','coords_ori','coords_merch','unix_time_conv','trans_datetime','dob_year',
                  'merch_lat','merch_long'])
```

```python
cc_6_encoded = pd.get_dummies(cc_6, columns=['merchant'], prefix='merch', drop_first=True)
cc_6_encoded = pd.get_dummies(cc_6_encoded, columns=['category'], prefix='cat', drop_first=True)
```

```python
cc_6_encoded.columns = ['_'.join(col.split()).lower() for col in cc_6_encoded.columns]
```

```python
feature_col = ['amt','lat','long','merch_lat','merch_long','is_fraud']
```

### Baseline Accuracy

```python
train[feature_col]['is_fraud'].value_counts(normalize=True)
```

```python
# for data modeling
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier

# for data evaluation
from sklearn.metrics import confusion_matrix, plot_confusion_matrix, accuracy_score, precision_score, \
                            recall_score, f1_score, plot_roc_curve, roc_auc_score
```

### Train-Test Split

```python
X = train[feature_col].drop(columns= 'is_fraud')
y = train['is_fraud']
```

```python
# split the datasets into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.30,
                                                    random_state=42,
                                                    stratify=y)
```

```python
# check shape of the train and test sets
print(f'X_train shape: {X_train.shape}')
print(f'X_test shape: {X_test.shape}')
print(f'y_train shape: {y_train.shape}')
print(f'y_test shape: {y_test.shape}')
```

### Logistic Regression

```python
# Instantiate our model.
lr = LogisticRegression()
```

```python
# Fit our model.
lr.fit(X_train, y_train)
```

```python
# Evaluate model.
lr.score(X_train, y_train)
```

```python
# Evaluate model.
lr.score(X_test, y_test)
```

```python
# plot confusion matrix
plot_confusion_matrix(lr, X_test, y_test, display_labels=['not_fraud', 'is_fraud'], 
                      cmap='Blues', values_format='d');
```

```python
# print summarize results
y_pred = lr.predict(X_test)

print(f"====== {lr}'s Metrics ======")
print(f"Train Score: {lr.score(X_train, y_train)}")
print(f"Test Score: {lr.score(X_test, y_test)}")    
print(f"Precision Score: {precision_score(y_test, y_pred)}")
print(f"Recall Score: {recall_score(y_test, y_pred)}")
print(f"f1-Score: {f1_score(y_test, y_pred)}")
print(f"roc_auc Score: {roc_auc_score(y_test, lr.predict_proba(X_test)[:,1])}")
```
